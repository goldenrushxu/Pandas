```python
#how to handle strings
import pandas as pd
```

```python
fpath = 'tianqi.csv'
```

```python
df = pd.read_csv(fpath)
```

```python
df
```

```python
condition = df['YMD'].str.endswith('01/2018')
condition
```

```python
df[condition].head(8)
```

```python
df.loc[df['YMD'].str.endswith('01/2018'),'Fengli'].head(80)
```

```python
df.loc[:,'bWendu'] = df['bWendu'].str.replace('°C','').astype(int)
df.loc[:,'yWendu'] = df['yWendu'].str.replace('°C','').astype(int)
df.loc[:,'YMD'] =  df['YMD'].str.zfill(10)
df
```

```python
df.loc[:,'YMD'] = df['YMD'].str.replace('/','').str[-4:] + df['YMD'].str.replace('/','').str[-6:-4] + df['YMD'].str.replace('/','').str[:-6] 
df
```

```python
df.loc[:,'中文日期'] = df['YMD'].str[:4] + '年' + df['YMD'].str[4:6] + '月' + df['YMD'].str[6:] + '日'
df
```

```python
df.loc[:,'中文日期'] = df['中文日期'].str.replace('年','').str.replace('月','').str.replace('日','')           #use individial replace functions
# df.loc[:,'中文日期'] = df['中文日期'].str.replace('[年月日]','')             #use regulator expression
df
```
