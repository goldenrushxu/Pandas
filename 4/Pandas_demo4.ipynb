```python
import pandas as pd
```

```python
fpath = "tianqi.csv"
df = pd.read_csv(fpath)
```

```python
df.head()
```

replace all the temperature into int data type to calculate

```python
df.loc[:, 'bWendu'] = df['bWendu'].str.replace("°C","").astype('int32')
df.loc[:, 'yWendu'] = df['yWendu'].str.replace("°C","").astype('int32')
```

```python
df.head() 
```

add a new column


```python
df.loc[:,'Wencha'] = df['bWendu']-df['yWendu'] #df['bWendu']is a series, it minus with df['yWendu'] series according to index one by one
```

```python
df.head()
```

Use apply() to add a new column

```python
def get_wendu_type(x):
    if x['bWendu']>33:
        return '高温'
    if x['yWendu']<-10:
        return '低温'
    return '常温'
```

```python
df.loc[:,'Wendu_type'] = df.apply(get_wendu_type, axis=1)
```
