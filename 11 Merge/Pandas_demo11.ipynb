```python
import pandas as pd
fpath_ratings = 'ratings.csv'
fpath_movies = 'movies.csv'
fpath_links = 'links.csv'
fpath_tags = 'tags.csv'
```

```python
df_ratings = pd.read_csv(fpath_ratings)
df_ratings.head()
```

```python
df_movies = pd.read_csv(fpath_movies)
df_movies.head()
```

```python
df_links = pd.read_csv(fpath_links)
# df_links.head()
```

```python
df_tags = pd.read_csv(fpath_tags)
df_tags.head()
```

```python
df_ratings_movies = pd.merge(
    df_ratings,df_movies,left_on='movieId',right_on='movieId',how='inner'
)
df_ratings_movies.loc[df_ratings_movies['movieId'] == 50]
```

```python
df_ratings_movies_tags = pd.merge(
    df_ratings_movies, df_tags, left_on='movieId', right_on= 'movieId', how='inner'
)
df_ratings_movies_tags.loc[df_ratings_movies_tags['movieId'] == 50]
```
